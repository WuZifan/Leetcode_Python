    1、看思路大家都一样的，都是从大到小遍历回文数
    2、然后对每个回文数，看是否能够被n位数整除，且得到另一个n位数。



```python
class Solution:
    def largestPalindrome(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n==1:
            return 9
        max_num='9'*n
        for j in range(1,2*n+1)[::-1]:
            print(j)
            if j%2==1:
                temp_len=int((j+1)/2)
                start_num='9'*temp_len
                for i in range(int(start_num)+1):
                    temp_pali=int(start_num[:-1]+start_num[::-1])
                    for k in range(10**(n-1),int(temp_pali/(10**(n-1)))+1)[::-1]:
                        another_part=temp_pali/k
                        if temp_pali%k==0 and len(str(int(another_part)))==n:
                            return temp_pali%1337,temp_pali
                    
                    start_num=str(int(start_num)-1)
                    if len(start_num)<temp_len:
                        break
            else:
                temp_len=int((j+1)/2)
                start_num='9'*temp_len
                for i in range(int(start_num)+1):
                    temp_pali=int(start_num+start_num[::-1])
                    for k in range(10**(n-1),int('9'*n)+1)[::-1]:
                        another_part=temp_pali/k
                        if temp_pali%k==0 and len(str(int(another_part)))==n:
                            return temp_pali%1337,temp_pali
                    start_num=str(int(start_num)-1)
                    if len(start_num)<temp_len:
                            break
        
        
```
