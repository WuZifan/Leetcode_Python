# 881救生船思路：
题目描述：

    1、船上最多两个人。
    2、人的总重量之和不能超过船的载重
    3、给定一些人的重量，问最少几艘船。

思路：

    1、因为限定了是2个人，那么就是以下条件：
        1.1 从最胖的开始，上完最胖的后，能不能上一个最轻的。
        1.2 如果可以上，那么两人一起走
        1.3 如果不行，那么最胖的先走，第二胖的继续重复1.1~1.3过程

大家的思路大体都是这样，具体实操起来：

    1、要多重量的list进行排序。
    2、尽可能不要操作list，用游标或者指针表示现在胖子上到哪里，瘦子上到哪里了。
    3、由于这里是最多两个人，其实简化了约束，如果是三个人，那么就比较复杂了。

```python
class Solution:
    def numRescueBoats(self, people, limit):
        """
        :type people: List[int]
        :type limit: int
        :rtype: int
        """
        # 从小到大
        people.sort()
        cnt=0
        while len(people)>0:
            fat=people.pop(-1)
            temp_in=fat
            cnt+=1
            num_per=1
            while len(people)>0 and num_per<2 and people[0]<=limit-temp_in:
                temp_in+=people.pop(0)
                num_per+=1
        return cnt
    
    # 这里相当于用指针替代了我之前写的pop方法。
    # 其实这不失为在解决这些问题时候的思路，即list问题的时候，从指针，即下标的角度去考虑
    # 而不拘泥于对list本身的操作。
    def numRescueBoats2(self,people,limit):
        people.sort()
        i, j, s = 0,len(people)-1, 0
        while i <= j:
            s += 1 
            print(people[i] + people[j] <= limit)
            # False 表示这艘船只能一个人上，不能让最小的上
            # True 表示这艘船可以两个人一起上所以i可以往前加
            i += people[i] + people[j] <= limit
            j -= 1
        return s
```
